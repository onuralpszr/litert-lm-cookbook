# Example 03: System Prompt

Seed the conversation with a system message before the first user turn.
The model adopts the persona described in the system message and keeps it across all subsequent turns.

In this example the model is told to act as a concise Python expert.

```python
!pip install litert-lm-api-nightly litert-lm -q
```

```python
import subprocess
subprocess.run([
    "curl", "-L",
    "https://huggingface.co/litert-community/gemma-4-E2B-it-litert-lm/resolve/main/gemma-4-E2B-it.litertlm?download=true",
    "-o", "/content/gemma-4-E2B-it.litertlm"
], check=True)
```

```python
import litert_lm

MODEL_PATH = "/content/gemma-4-E2B-it.litertlm"

litert_lm.set_min_log_severity(litert_lm.LogSeverity.ERROR)

SYSTEM_PROMPT = (
    "You are a concise Python expert. "
    "Always respond with short, clear explanations and include a code example when relevant."
)

messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": SYSTEM_PROMPT}],
    }
]

with litert_lm.Engine(MODEL_PATH) as engine:
    with engine.create_conversation(messages=messages) as conversation:
        questions = [
            "What is a list comprehension?",
            "How do I read a file line by line?",
        ]
        for q in questions:
            print(f"Q: {q}")
            response = conversation.send_message(q)
            print(f"A: {response['content'][0]['text']}\n")
```

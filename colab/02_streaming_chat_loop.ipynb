# Example 02: Streaming Chat Loop

Tokens are printed as they are generated rather than waiting for the full response.

Colab does not support an interactive terminal loop, so this notebook uses a fixed list of questions.
Edit the `questions` list in the last cell to try your own prompts.

```python
import subprocess
subprocess.run([
    "curl", "-L",
    "https://huggingface.co/litert-community/gemma-4-E2B-it-litert-lm/resolve/main/gemma-4-E2B-it.litertlm?download=true",
    "-o", "/content/gemma-4-E2B-it.litertlm"
], check=True)
```

```python
import litert_lm

MODEL_PATH = "/content/gemma-4-E2B-it.litertlm"

litert_lm.set_min_log_severity(litert_lm.LogSeverity.ERROR)

questions = [
    "What is on-device AI?",
    "Name three countries on the Black Sea coast.",
    "Explain what a transformer model is in two sentences.",
]

with litert_lm.Engine(MODEL_PATH) as engine:
    with engine.create_conversation() as conversation:
        for question in questions:
            print(f"Q: {question}")
            print("A: ", end="", flush=True)
            for chunk in conversation.send_message_async(question):
                for item in chunk.get("content", []):
                    if item.get("type") == "text":
                        print(item["text"], end="", flush=True)
            print("\n")
```

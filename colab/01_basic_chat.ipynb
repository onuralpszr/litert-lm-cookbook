# Example 01: Basic Chat

Send a single message to Gemma-4 E2B and print the full response.
This is the simplest possible usage of the LiteRT-LM Python API.

Run the two setup cells first, then the chat cell.

```python
# Install dependencies
!pip install litert-lm-api-nightly litert-lm -q
```

```python
# Download the model (this takes a few minutes the first time)
import subprocess
subprocess.run([
    "curl", "-L",
    "https://huggingface.co/litert-community/gemma-4-E2B-it-litert-lm/resolve/main/gemma-4-E2B-it.litertlm?download=true",
    "-o", "/content/gemma-4-E2B-it.litertlm"
], check=True)
```

```python
import litert_lm

MODEL_PATH = "/content/gemma-4-E2B-it.litertlm"

litert_lm.set_min_log_severity(litert_lm.LogSeverity.ERROR)

with litert_lm.Engine(MODEL_PATH) as engine:
    with engine.create_conversation() as conversation:
        response = conversation.send_message("What is the capital of Turkey?")
        print(response["content"][0]["text"])
```

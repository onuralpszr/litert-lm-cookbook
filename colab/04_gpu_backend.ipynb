# Example 04: GPU Backend

Run inference on GPU instead of the default CPU backend.

Before running this notebook, make sure you have selected a GPU runtime:
go to **Runtime**, then **Change runtime type**, and select **T4 GPU** (or any other available GPU).

The first run compiles GPU kernels and stores them in `cache_dir`.
Subsequent runs reuse the cache and start up faster.

```python
import subprocess
subprocess.run([
    "curl", "-L",
    "https://huggingface.co/litert-community/gemma-4-E2B-it-litert-lm/resolve/main/gemma-4-E2B-it.litertlm?download=true",
    "-o", "/content/gemma-4-E2B-it.litertlm"
], check=True)
```

```python
# Verify GPU is available
!nvidia-smi
```

```python
import litert_lm

MODEL_PATH = "/content/gemma-4-E2B-it.litertlm"

litert_lm.set_min_log_severity(litert_lm.LogSeverity.ERROR)

with litert_lm.Engine(
    MODEL_PATH,
    backend=litert_lm.Backend.GPU,
    cache_dir="/tmp/litert-lm-cache",
) as engine:
    with engine.create_conversation() as conversation:
        response = conversation.send_message("Explain attention mechanism in transformers.")
        print(response["content"][0]["text"])
```
